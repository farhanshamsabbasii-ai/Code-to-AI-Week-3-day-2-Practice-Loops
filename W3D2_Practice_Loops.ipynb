### Exercise 2 — Countdown
Use a `while` loop to count down from 5 to 1, then print `"Blast off! 🚀"`.

````python
# Your code here
i = 5

while i >=1:
    print(i)
    if i == 1:
        print("Blast off! 🚀")
    i -= 1
````

### Exercise 5 — Skip the multiples of 3
Print numbers 1 to 20, but use `continue` to skip any number divisible by 3.

````python
# Your code here
for i in range(1, 21):
    if i % 3 == 0:
        continue
    print(i)
````

### Exercise 6 — Find the first big number
Given the list below, loop through it and print the **first** number greater than 50, then `break`.
```python
numbers = [12, 8, 45, 30, 67, 22, 90]
```

````python
numbers = [12, 8, 45, 30, 67, 22, 90]
# Your code here

for n in numbers:
    if n > 50:
        print(n)
        break
````

### Exercise 7 — Count the vowels
Given `word = "education"`, loop through each letter and count how many vowels it has.
*Hint: `if letter in "aeiou":`*

````python
word = "education"
# Your code here
count = 0

for v in word:
    if v in "aeiou":
        count +=1

print("Total number of vowels:" , count)
````
